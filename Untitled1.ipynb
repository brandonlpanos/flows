# Normalizing flows
Normalizing flows (NF) can transform simple probability distributions into more complex distributions via the application of a string of invertible and differentiable functions [(Rezende and Mohamed, 2016)](https://arxiv.org/abs/1505.05770). Manipulating a probability distribution into another probability distribution can be achieved via the change of variable formula, which requires the calculation of the Jacobian determinant. Each invertible transformation is carried out by a layer within a multilayer neural network. The modelâ€™s architecture has to be restrained such that the Jacobian becomes trivial to solve, i.e., (lower triangular, etc.). It turns out that the model becomes unconstrained when you take the limit, i.e., a continuous set of transformations. When this happens we achieve a new class of models called continuous normalizing flows. 

Here a neural network is used to model a time dynamic, meaning that the NN is now responsible for constructing a vector field that transports the probability mass distribution of a baseline distribution into the target distribution. The problem can be framed as solving an ordinary differential equation with initial conditions, hence the name Neural-ODE ([Chen et.al, 2019](https://arxiv.org/pdf/1806.07366.pdf)).

```python
import torch 
import pickle
import numpy as np
import torch.nn as nn
from sklearn.datasets import make_circles
from torch.utils.data import Dataset, DataLoader
from torchdiffeq import odeint_adjoint as odeint
from matplotlib.ticker import MultipleLocator
import matplotlib.pyplot as plt
```

```python
# Set the device to GPU since flows are computationally expensive
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

**Define the base and the target distributions** 

```python
# Simple Gaussian since is easy to sample and has a known density 
p_z0 = torch.distributions.MultivariateNormal(
    loc=torch.tensor([0.0, 0.0]).to(device),
    covariance_matrix=torch.tensor([[0.2, 0.0], [0.0, 0.2]]).to(device))
```

```python
def get_batch_targets(num_samples):
    points, _ = make_circles(n_samples=num_samples, noise=0.06, factor=0.5)
    x = torch.tensor(points).type(torch.float32).to(device)
    logp_diff_t1 = torch.zeros(num_samples, 1).type(torch.float32).to(device)
    return(x, logp_diff_t1)
```

**Creat a dataloader to generate target vectors**

```python
class CircleDataset(Dataset):
    def __init__(self, data):
        self.target = data[0]
        self.log_diff = data[1]
    def __len__(self):
        return len(self.target)
    def __getitem__(self, indx):
        return torch.Tensor(self.target[indx]), torch.Tensor(self.log_diff[indx])
targetloader = DataLoader(CircleDataset(get_batch_targets(12800)), batch_size=64, shuffle=True, pin_memory=True)
```

**Model the systems infinitesimal dynamics using a NN**   

```python
class CNF(nn.Module):
    def __init__(self, in_out_dim, hidden_dim, width):
        super().__init__()
        self.in_out_dim = in_out_dim
        self.hidden_dim = hidden_dim
        self.width = width
        self.hyper_net = HyperNetwork(in_out_dim, hidden_dim, width)

    def forward(self, t, states):
        z = states[0]
        logp_z = states[1]
        batchsize = z.shape[0]

        with torch.set_grad_enabled(True):
            z.requires_grad_(True)
            W, B, U = self.hyper_net(t)
            Z = torch.unsqueeze(z, 0).repeat(self.width, 1, 1)
            h = torch.tanh(torch.matmul(Z, W) + B)
            dz_dt = torch.matmul(h, U).mean(0)
            dlogp_z_dt = -trace_df_dz(dz_dt, z).view(batchsize, 1)

        return (dz_dt, dlogp_z_dt)
    
    
def trace_df_dz(f, z):
    sum_diag = 0.
    for i in range(z.shape[1]):
        sum_diag += torch.autograd.grad(f[:, i].sum(), z, create_graph=True)[0].contiguous()[:, i].contiguous()
    return sum_diag.contiguous()


class HyperNetwork(nn.Module):
    def __init__(self, in_out_dim, hidden_dim, width):
        super().__init__()

        blocksize = width * in_out_dim

        self.fc1 = nn.Linear(1, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        self.fc3 = nn.Linear(hidden_dim, 3 * blocksize + width)

        self.in_out_dim = in_out_dim
        self.hidden_dim = hidden_dim
        self.width = width
        self.blocksize = blocksize

    def forward(self, t):
        # predict params
        params = t.reshape(1, 1)
        params = torch.tanh(self.fc1(params))
        params = torch.tanh(self.fc2(params))
        params = self.fc3(params)

        # restructure
        params = params.reshape(-1)
        W = params[:self.blocksize].reshape(self.width, self.in_out_dim, 1)

        U = params[self.blocksize:2 * self.blocksize].reshape(self.width, 1, self.in_out_dim)

        G = params[2 * self.blocksize:3 * self.blocksize].reshape(self.width, 1, self.in_out_dim)
        U = U * torch.sigmoid(G)

        B = params[3 * self.blocksize:].reshape(self.width, 1, 1)
        return [W, B, U]
```

The idea in normalizing flows is to create flexible probability distributions over a continuous random variable $\mathbf{x}$ by sampling vectors $\mathbf{u}$ from some base distribution $q(\mathbf{u})$ and generating $\mathbf{x}$ via the application of a transformation $T: \mathbf{u}\to \mathbf{x}$, where $\mathbf{u}$ and $\mathbf{x}~\in\mathbb{R}^D$. Generally, the base distribution is simple to sample from with a known probability distribution. The problem is parameterized by the parameters of the base distribution $\phi$ and the parameters of the transformation $\psi$. The transformation can be modeled by a neural network $T_\psi$. This induces a family of distributions $p(\mathbf{x};\theta)$ over $\mathbf{x}$ where $\theta:=\{\phi,\psi\}$. A trandsfomation can be achieved via the change of variable formula, that is, we can obtain $p(\mathbf{x})$ form $q(\mathbf{u})$ via 

$$
p(x) = q(\mathbf{u})|\det J_T(\mathbf{u})|^{-1},
$$

where $J_T$ is the Jacobian matrix. This places a restriction on the transformation that it be both invertible and differentiable.  

Our goal is to transform our base distribution of noise $q(\mathbf{u})$ into the more complex target concentric circles distribution $p(\mathbf{x})$. Since we can sample from the target distribution but do not necessarily know the target density $p^*(\mathbf{x})$, the parameters $\theta$ can be optimized via the minimization of the forward $KL$-divergance

$$
\begin{aligned}
\mathcal{L} & = D_{KL}[p^*(\mathbf{x})||p(\mathbf{x},\theta)]\\
& =-\mathbb{E}_{p_{\mathbf{x}}^*(\mathbf{x})}\left[\log q\left(T^{-1}(\mathbf{x} ; \phi) , \psi\right)+\log~\left|\operatorname{det} J_{T^{-1}}(\mathbf{x} ; \phi)\right|\right]+\text { const. }
\end{aligned}
$$

This means that in order to optimize the flow we need to implement the reverse transformation $T^{-1}$ and maximize the flowâ€™s output $T^{-1}| \mathbf{x}\to\mathbf{u}$ under $q$. The expectation value can also be performed via Montecarlo in practice.  

The theoretical advantage of modeling the flow as a continuous time dynamic is that the Jacobian matrix becomes free-flowing, meaning one no longer has to restrict the network to make the calculation of the Jacobian determinant computationally efficient. In this framework, the systemâ€™s infinitesimal dynamics are modeled 

$$
\frac{d\mathbf{z}_t}{dt} = g_\phi(t,\mathbf{z}_t), 
$$

where the only restriction is that $\mathbf{z_t}$ be uniform Lipschitz continiouse. Both the forward transformation and the log density can be evaluated by integration 

$$
\left[\begin{array}{c}
\mathbf{x} \\
\log p(\mathbf{x})
\end{array}\right]=\left[\begin{array}{c}
\mathbf{u} \\
\log q(\mathbf{u})
\end{array}\right]+\int_{t=t_0}^{t_1}\left[\begin{array}{c}
g_\phi\left(t, \mathbf{z}_t\right) \\
-\operatorname{Tr}\left\{J_{g_\phi(t,)}\left(\mathbf{z}_t\right)\right\}
\end{array}\right] d t
$$  

The initial value problem can be solved via the use of a black box ordinary differential equations (ODE) solver. Backpropagation is done by deriving the sensitivity of the loss function on the modelâ€™s weights $\phi$ via the use of the adjoint sensitivity method, where a second differential equation is solved backward in time such that the integration results in the partial derivative of the loss function with respect to all model parameters 

$$
\frac{\partial \mathcal{L}(\mathbf{x} ; \phi)}{\partial \phi}=\int_{t=t_1}^{t_0} \frac{\partial \mathcal{L}}{\partial \mathbf{z}_t} \frac{\partial g_\phi\left(t, \mathbf{z}_t\right)}{\partial \phi} d t
$$

There is a PyTorch implementation of such a solver that can be found [here](https://github.com/rtqichen/torchdiffeq).

```python
# Define the training loop
def train(model, optimizer, n_epochs, batch_size):
    best_loss = 1e10
    num_samples = 64
    for epoch in range(n_epochs):
        epoch_loss = 0
        for x, logp_diff_t1 in targetloader:
            optimizer.zero_grad()
            
            z_t, logp_diff_t = odeint(
                model,
                (x, logp_diff_t1),
                torch.tensor([t1, t0]).type(torch.float32).to(device),
                atol=1e-5,
                rtol=1e-5,
                method='dopri5',
            )

            z_t0, logp_diff_t0 = z_t[-1], logp_diff_t[-1]
            
            logp_x = p_z0.log_prob(z_t0).to(device) - logp_diff_t0.view(-1)
            loss = -logp_x.mean(0)
            
            loss.backward()
            optimizer.step()
            
            epoch_loss += loss.item()

        print(f"Epoch {epoch+1} loss: {epoch_loss/len(targetloader)*batch_size}")
        if epoch_loss/len(targetloader)*batch_size < best_loss:
            best_loss = epoch_loss/len(targetloader)*batch_size
            torch.save(model.state_dict(), "best_model_im.pth")
```

```python
# Time grid
t0 = 0
t1 = 1
# Initialize the model and optimizer
ode_func = CNF(in_out_dim=2, hidden_dim=32, width=64).to(device)
optimizer = torch.optim.Adam(ode_func.parameters(), lr=1e-4)
# Train the model
train(ode_func, optimizer, n_epochs=350, batch_size=64)
```

**Evaluate flow**

```python
# Here we integrate forwards from t0 --> t1 and go from noise to target z0 --> z1

viz_samples = 80000
viz_timesteps = 100
z_t0 = p_z0.sample([viz_samples]).to(device)
logp_diff_t0 = torch.zeros(viz_samples, 1).type(torch.float32).to(device)

z_t_samples, _ = odeint(
    ode_func,
    (z_t0, logp_diff_t0),
    torch.tensor(np.linspace(t0, t1, viz_timesteps)).to(device),
    atol=1e-5,
    rtol=1e-5,
    method='dopri5',
)

z_t_samples = z_t_samples.detach().numpy()
```
